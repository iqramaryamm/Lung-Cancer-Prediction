```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
df=pd.read_excel("cancer_patient_data_sets.xlsx")
```

```python

df.head()
```

```python
df.head()
```

```python

new_df=df.loc[:,df.columns[2:]]
new_df
```

```python
len(new_df.columns)
print(df.shape)
```

```python
input_df=new_df.iloc[:,:-1]
# output_df=df[-1]
output_df=new_df.iloc[:,-1]
```

```python
print(len(df[df['Level']=='Low']))
print(len(df[df['Level']=='High']))
print(len(df[df['Level']=='Medium']))

print(df['Level'].isnull().sum())
```

```python
df.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_df,output_df,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
```

```python
y_pred = RF.predict(x_test)
# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.svm import SVC
SVC_model=SVC()
SVC_model.fit(x_train,y_train)
```

```python
y_pred2=SVC_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred2)
print(f'Accuracy: {accuracy:.2f}')
```
